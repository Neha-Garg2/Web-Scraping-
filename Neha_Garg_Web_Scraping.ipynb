# **Web Scraping**

**IMPORT LIABRARIES**

```
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

**Request URL**

```
page = requests.get('https://www.amazon.in/mobile-phones/b/?ie=UTF8&node=1389401031&ref_=nav_cs_mobiles')
```

```
soup = BeautifulSoup(page.text, 'html.parser')
```

**Find content of webpage** &&
**To print all "<a" tags**

```
m = soup.find_all('a')
m
```

**To save content**

```
links = set()

for l in m:
    if (l.get('href') != '#'):
      lText = "https://www.amazon.in/mobile-phones/b/?ie=UTF8&node=1389401031&ref_=nav_cs_mobiles" + str(l.get('href'))
      links.add(l)
      print(lText)
```
